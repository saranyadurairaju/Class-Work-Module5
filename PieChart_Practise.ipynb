```python
import matplotlib.pyplot as plt

ages = [17, 18, 19, 20, 21]

total_population = [21217467, 27958147, 20859088, 28882735, 19978972]
explode = [0, 0, 0, 0, 0.1]

plt.pie(total_population, labels=ages, explode=explode,
        shadow=True,
        autopct='%1.1f%%',
        wedgeprops={'edgecolor': 'black'})
plt.title("A Simple Pie Chart")
plt.show()
```
